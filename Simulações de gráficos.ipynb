```python
from quilt.data.alifersales import burj20141_processed
import pandas as pd
import matplotlib.pyplot as plt
import random as r
df = burj20141_processed()
```

# Jair Bolsonaro e Jean WIllys

```python
pai = df[df['numero_votavel']== 1120]
jean = df[df['numero_votavel']== 5005]
conj = pd.merge(pai,jean, on=['numero_zona', 'numero_secao'], suffixes=['_pai','_jean'])
conjx = conj[['qtde_votos_pai','qtde_votos_jean']]
```

```python
A = []
C = []
for x in conjx.as_matrix():
    B = []
    for y in x:
        B.append(y)
    A.append(B[0])
    C.append(B[1])
```

```python
plt.scatter(A,C)
plt.xlim(0,140)
plt.ylim(0,140)
plt.xlabel('Jair Bolsonaro')
plt.ylabel('Jean Wyllys')
```

# Função

```python
def doisdeputados(a,b, df, nome_a= 'Candidato A', nome_b='Candidato B', multiplicador = 1 ):
    pai = df[df['numero_votavel']== a]
    jean = df[df['numero_votavel']== b]
    conj = pd.merge(pai,jean, on=['numero_zona', 'numero_secao'], suffixes=['_a','_b'], how='outer')
    conjx = conj[['qtde_votos_a','qtde_votos_b']].fillna(0)
    A = []
    C = []
    S = []
    K = []
    for x in conjx.as_matrix():
        B = []
        for y in x:
            B.append(y)
        A.append(B[0]+r.normalvariate(0,0.2))
        C.append(B[1]+r.normalvariate(0,0.2))
        K.append(B[0]+B[1])
    plt.scatter(A,C, s=0.3, c=K)
    plt.xlim(-2,multiplicador*80)
    plt.ylim(-2,multiplicador*80)
    plt.xlabel(nome_a)
    plt.ylabel(nome_b)
    plt.show()
```
